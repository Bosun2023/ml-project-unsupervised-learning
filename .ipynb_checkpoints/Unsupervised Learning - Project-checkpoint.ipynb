# Part II - KMeans Clustering

The objective of the analysis is to group similar products together into clusters based on their attributes such as fresh, milk, grocery, frozen, detergents_paper, and delicatessen. To perform the k-means clustering analysis, you will need to pre-process the dataset, determine the optimal number of clusters, initialize the centroids, assign data points to clusters, update the centroids, and repeat until convergence.

```python
# Data Pre-processing
from sklearn.preprocessing import StandardScaler

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(wholesale_data[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']])
```

```python
# Determine the Optimal Number of Clusters
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Determine the optimal number of clusters using the Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

# Determine the optimal number of clusters using the Silhouette Score
silhouette_scores = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    cluster_labels = kmeans.fit_predict(scaled_data)
    silhouette_avg = silhouette_score(scaled_data, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plot the Silhouette Score
plt.figure(figsize=(8, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='--')
plt.title('Silhouette Score for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
import os

# Set the OMP_NUM_THREADS environment variable
os.environ['OMP_NUM_THREADS'] = '2'

# Initialize and fit K-Means with the optimal number of clusters
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
cluster_labels = kmeans.fit_predict(scaled_data)
```

```python
# Analyze the Clusters
# Add cluster labels to the original dataset
wholesale_data['Cluster'] = cluster_labels

# Explore cluster characteristics
cluster_means = wholesale_data.groupby('Cluster').mean()
print(cluster_means)
```

```python
# Visualize cluster
# Scatter plot for 'Fresh' and 'Milk'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=wholesale_data, x='Fresh', y='Milk', hue='Cluster', palette='viridis')
plt.title('Scatter Plot: Fresh vs. Milk by Cluster')
plt.show()
```

```python
# Bar plot to visualize the distribution of data points across clusters
plt.figure(figsize=(8, 5))
sns.countplot(data=wholesale_data, x='Cluster', palette='viridis')
plt.title('Cluster Distribution')
plt.xlabel('Cluster')
plt.ylabel('Count')
plt.show()
```

```python
# Box plots for each feature, grouped by cluster, to see how each cluster differs in terms of spending patterns.
plt.figure(figsize=(12, 6))
sns.boxplot(data=wholesale_data, x='Cluster', y='Fresh', palette='viridis')
plt.title('Fresh Spending by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Fresh')
plt.show()
```

# Part III - Hierarchical Clustering 

Hierarchical clustering is a popular unsupervised machine learning algorithm that is used to identify patterns and group similar data points together in a hierarchy. The algorithm works by iteratively merging or splitting clusters based on a similarity measure until a dendrogram is formed.

To perform hierarchical clustering analysis, you will need to pre-process the dataset, determine the optimal number of clusters using techniques such as dendrogram.

```python
#  Data Pre-processing
from sklearn.preprocessing import StandardScaler

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(wholesale_data[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']])
```

```python
# Determine the Optimal Number of Clusters
import scipy.cluster.hierarchy as sch

# Perform hierarchical clustering
dendrogram = sch.dendrogram(sch.linkage(scaled_data, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distances')
plt.show()
```

```python
# Hierarchical Clustering
from sklearn.cluster import AgglomerativeClustering

# Create an AgglomerativeClustering model with the chosen number of clusters
num_clusters = 3  
hierarchical_cluster = AgglomerativeClustering(n_clusters=num_clusters, affinity='euclidean', linkage='ward')

# Fit the model to the scaled data and get cluster labels
cluster_labels = hierarchical_cluster.fit_predict(scaled_data)
```

```python
# Analyze the Clusters
# Add cluster labels to the original dataset
wholesale_data['Cluster'] = cluster_labels

# Explore cluster characteristics
cluster_means = wholesale_data.groupby('Cluster').mean()
print(cluster_means)
```

# Part IV - PCA

In this section you are going to perform principal component analysis (PCA) to draw conclusions about the underlying structure of the wholesale customer data. Since using PCA on a dataset calculates the dimensions which best maximize variance, we will find which compound combinations of features best describe customers.

```python
# Data Pre-processing
# Standardize the data (if not done already)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(wholesale_data[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']])
```

```python
#  Perform PCA
from sklearn.decomposition import PCA

# Create a PCA instance with the desired number of components (e.g., 6 for all features)
num_components = 6
pca = PCA(n_components=num_components)

# Fit PCA to the scaled data and transform it
pca_data = pca.fit_transform(scaled_data)
```

```python
# Explained Variance
explained_variance = pca.explained_variance_ratio_
cumulative_variance = explained_variance.cumsum()

# Print explained variance for each component and cumulative variance
print("Explained Variance Ratio:")
print(explained_variance)
print("\nCumulative Explained Variance:")
print(cumulative_variance)
```

```python
# visualize the explained variance using a scree plot to help determine the number of principal components to retain
# Plot explained variance ratio
plt.figure(figsize=(8, 6))
plt.bar(range(1, num_components + 1), explained_variance, align='center', label='Explained Variance')
plt.step(range(1, num_components + 1), cumulative_variance, where='mid', label='Cumulative Explained Variance')
plt.xlabel('Principal Components')
plt.ylabel('Variance Explained')
plt.legend()
plt.title('Scree Plot')
plt.show()
```

Note: In the scree plot, the "elbow" point where adding more principal components does not significantly increase the explained variance is at principal component 3. This point can help  decide the number of components to retain.

```python
# To interpret principal components, examine the loadings (coefficients) of the original features on each principal component to interpret their meaning

# Get the loadings of the original features on each principal component
loadings = pca.components_

# Create a DataFrame to display loadings
loadings_df = pd.DataFrame(loadings, columns=['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen'])
print("Loadings of Features on Principal Components:")
print(loadings_df)
```
