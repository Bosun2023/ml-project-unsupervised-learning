- **Tail Direction**: A left-skewed (positively skewed) histogram has a longer tail on the left side and a shorter tail on the right side.
  
- **Central Tendency**: The peak of the histogram is on the right side, indicating that the central tendency (mean or median) is likely greater than the mode.
  
- **Outliers**: Extreme values, if present, are typically on the right side and can significantly affect the mean.
  
- **Data Spread**: Data is spread out towards lower values, resulting in a wider range compared to a symmetric distribution.
  
- **Interpretation**: Left-skewed distributions often indicate that most observations have lower values, with a few extreme values on the right. This may suggest scenarios with many small transactions or low quantities of products.
  
- **Statistical Measures**: The mean is greater than the median due to the influence of the right tail.
  
- **Data Transformation**: Left-skewed data may benefit from transformations like log transformation to reduce skewness and make the data more symmetric.

# Part II - KMeans Clustering

The objective of the analysis is to group similar products together into clusters based on their attributes such as fresh, milk, grocery, frozen, detergents_paper, and delicatessen. To perform the k-means clustering analysis, you will need to pre-process the dataset, determine the optimal number of clusters, initialize the centroids, assign data points to clusters, update the centroids, and repeat until convergence.

```python
# Data Pre-processing
from sklearn.preprocessing import StandardScaler

# Select the log-transformed columns
log_transformed_columns = ['Log_Fresh', 'Log_Milk', 'Log_Grocery', 'Log_Frozen', 'Log_Detergents_Paper', 'Log_Delicassen']

# Initialize the StandardScaler
scaler = StandardScaler()

# Standardize the log-transformed data
scaled_log_data = scaler.fit_transform(wholesale_data[log_transformed_columns])
```

```python
# Determine the Optimal Number of Clusters
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Determine the optimal number of clusters using the Elbow Method
wcss_log = []
for i in range(1, 11):
    kmeans_log = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans_log.fit(scaled_log_data)
    wcss_log.append(kmeans_log.inertia_)

# Plot the Elbow Method for log-transformed data
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss_log, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal Number of Clusters (Log-Transformed Data)')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

# Determine the optimal number of clusters using the Silhouette Score for log-transformed data
silhouette_scores_log = []
for i in range(2, 11):
    kmeans_log = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    cluster_labels_log = kmeans_log.fit_predict(scaled_log_data)
    silhouette_avg_log = silhouette_score(scaled_log_data, cluster_labels_log)
    silhouette_scores_log.append(silhouette_avg_log)

# Plot the Silhouette Score for log-transformed data
plt.figure(figsize=(8, 6))
plt.plot(range(2, 11), silhouette_scores_log, marker='o', linestyle='--')
plt.title('Silhouette Score for Optimal Number of Clusters (Log-Transformed Data)')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()
```


Optimal Number of Clusters
After performing clustering analysis on the log-transformed data, two different methods were used to determine the optimal number of clusters:

Elbow Method: The Elbow Method involves plotting the Within-Cluster Sum of Squares (WCSS) for a range of cluster numbers and looking for the "elbow" point where the rate of decrease in WCSS starts to slow down. In our analysis, the Elbow Method indicated that the optimal number of clusters is 2, as there is an evident bend in the plot at this point.

Silhouette Score: The Silhouette Score measures how similar each data point in one cluster is to the data points in the neighboring clusters. It ranges from -1 (a poor clustering) to +1 (a perfect clustering). A higher Silhouette Score indicates better-defined clusters. Our analysis showed that the highest Silhouette Score was achieved with 2 clusters, further supporting the choice of 2 clusters.

Both methods consistently suggest that dividing the data into 2 clusters is appropriate for this analysis. Therefore, we will proceed with KMeans clustering using 2 clusters for the log-transformed data.

```python
import os

# Set the OMP_NUM_THREADS environment variable
os.environ['OMP_NUM_THREADS'] = '2'

# Initialize and fit K-Means with the optimal number of clusters
from sklearn.cluster import KMeans

# Initialize K-Means with 2 clusters
kmeans = KMeans(n_clusters=2, init='k-means++', max_iter=300, n_init=10, random_state=0)

# Fit K-Means to the scaled log-transformed data
cluster_labels = kmeans.fit_predict(scaled_log_data)
```

```python
# Analyze the Clusters
# Add cluster labels to the original dataset
wholesale_data['Cluster'] = cluster_labels

# Explore cluster characteristics
cluster_means = wholesale_data.groupby('Cluster').mean()
print(cluster_means)
```

# Part III - Hierarchical Clustering 

Hierarchical clustering is a popular unsupervised machine learning algorithm that is used to identify patterns and group similar data points together in a hierarchy. The algorithm works by iteratively merging or splitting clusters based on a similarity measure until a dendrogram is formed.

To perform hierarchical clustering analysis, you will need to pre-process the dataset, determine the optimal number of clusters using techniques such as dendrogram.

```python
# Data Pre-processing
from sklearn.preprocessing import StandardScaler

# Select the log-transformed columns
log_transformed_columns = ['Log_Fresh', 'Log_Milk', 'Log_Grocery', 'Log_Frozen', 'Log_Detergents_Paper', 'Log_Delicassen']

# Initialize the StandardScaler
scaler = StandardScaler()

# Standardize the log-transformed data
scaled_log_data = scaler.fit_transform(wholesale_data[log_transformed_columns])
```

```python
# Determine the Optimal Number of Clusters
import scipy.cluster.hierarchy as sch
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import pdist

# Calculate the linkage matrix using Ward's method
linkage_matrix = sch.linkage(scaled_log_data, method='ward')
```

```python
# Plot the dendrogram to visualize the hierarchical clustering
plt.figure(figsize=(12, 6))
dendrogram = sch.dendrogram(linkage_matrix)
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

# Interpretation of Hierarchical Clustering Results

After performing hierarchical clustering on the log-transformed and standardized wholesale customer data, I analyzed the dendrogram to determine the optimal number of clusters. The dendrogram visually represents the hierarchy of data points and clusters, with the vertical lines indicating where clusters are merged.

Upon careful examination of the dendrogram, I observed that it exhibits a clear separation into two distinct branches or subtrees. This separation suggests that partitioning the data into two clusters is a suitable choice for our analysis. The heights at which clusters are merged in the dendrogram support the idea of having two clusters, as there are no other obvious branches or clusters that form higher in the hierarchy.

Therefore, based on the dendrogram's structure and the clear separation into two branches, I have chosen to proceed with hierarchical clustering using two clusters for further analysis. This decision aligns with the inherent structure of the data and will help to explore meaningful patterns and differences within the dataset.

# Part IV - PCA

In this section you are going to perform principal component analysis (PCA) to draw conclusions about the underlying structure of the wholesale customer data. Since using PCA on a dataset calculates the dimensions which best maximize variance, we will find which compound combinations of features best describe customers.

```python
# Data Pre-processing
# Standardize the data
from sklearn.preprocessing import StandardScaler

# Select the log-transformed columns
log_transformed_columns = ['Log_Fresh', 'Log_Milk', 'Log_Grocery', 'Log_Frozen', 'Log_Detergents_Paper', 'Log_Delicassen']

# Initialize the StandardScaler
scaler = StandardScaler()

# Standardize the log-transformed data
scaled_log_data = scaler.fit_transform(wholesale_data[log_transformed_columns])
```

```python
#  Perform PCA
from sklearn.decomposition import PCA

# Initialize PCA with the desired number of components 
num_components = 6
pca = PCA(n_components=num_components)

# Fit PCA to the scaled data
pca.fit(scaled_log_data)

# Transform the data into principal components
pca_data = pca.transform(scaled_log_data)

# Create a DataFrame to store the PCA results
pca_df = pd.DataFrame(data=pca_data, columns=[f'PC{i}' for i in range(1, num_components + 1)])
```

```python
# Explained variance ratios
explained_variance_ratios = pca.explained_variance_ratio_
cumulative_variance = explained_variance_ratios.cumsum()

# Print the explained variance ratios for each component and cumulative variance
for i, explained_variance_ratio in enumerate(explained_variance_ratios, 1):
    print(f'Explained Variance (PC{i}): {explained_variance_ratio:.2%}')
    print(f'Cumulative Explained Variance (PC{i}): {cumulative_variance[i-1]:.2%}')
```

```python
import matplotlib.pyplot as plt

# Create a scree plot
plt.figure(figsize=(8, 6))
plt.bar(range(1, len(explained_variance_ratios) + 1), explained_variance_ratios, label='Explained Variance Ratio', align='center', alpha=0.8)
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--', label='Cumulative Explained Variance', color='orange')
plt.xlabel('Principal Components')
plt.ylabel('Variance Explained')
plt.legend()
plt.title('Scree Plot')
plt.grid(True)
plt.show()
```

# Scree Plot for Principal Component Analysis

In the scree plot above, I visualize the explained variance ratio for each principal component to determine the number of components to retain in the analysis. The plot displays two important metrics:

Explained Variance: Each bar on the plot represents the proportion of variance explained by a single principal component. Moving from left to right, the explained variance decreases for each successive component.

Cumulative Explained Variance: The cumulative explained variance, shown as a step plot, illustrates the total variance explained as more principal components are included. It quantifies how much of the dataset's variability have been captured by retaining a given number of components.

In this specific scree plot, the "elbow" point, which indicates a diminishing return in terms of explained variance, is observed at principal component 2. Beyond this point, adding more components provides relatively little increase in explained variance. Therefore, to strike a balance between retaining information and reducing dimensionality, I have chosen to retain the first two principal components, which together explain a substantial portion of the variance in our data.

Retaining these two principal components will help to analyze and understand the underlying structure of the wholesale customer data efficiently.

```python
# To interpret principal components, examine the loadings (coefficients) of the original features on each principal component to interpret their meaning

# Get the loadings of the log-transformed features on each principal component
loadings = pca.components_

# Create a DataFrame to display loadings
loadings_df = pd.DataFrame(loadings, columns=log_transformed_columns)
print("Loadings of Log-Transformed Features on Principal Components:")
print(loadings_df)

```
